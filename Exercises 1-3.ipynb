```python
target_name = "ESTEBAN"
SEED_NUM = 8
redacted_name = "" 

for char in target_name:
    if char in "AEIOU":
        redacted_name += str(SEED_NUM)  
    else:
        redacted_name += char           

print("Original:", target_name)
print("Seed Used:", SEED_NUM)
print("Redacted:", redacted_name)
```

```python
engineer_name = "ANDREI"
first_letter = engineer_name[0]
ASCII = ord(first_letter)

loops = 10

verification_hash = (ASCII * SEED_NUM) - loops

print("First Letter ASCII:", ASCII)
print("Seed:", SEED_NUM)
print("Total Loops:", loops)
print("Verification Hash:", verification_hash)
```

```python
correct_password = "jadeee"
attempts = 0

while True:
    user_input = input("Enter password: ")
    attempts += 1

    if user_input == correct_password:
        print("Access Granted!")
        break
    else:
        print("Incorrect. Try again.")

        print("Attempts before success:", attempts)
```
